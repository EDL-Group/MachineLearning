# Multicollinearity
This jupyter notebook is focused on how to easily visualise correlation values to spot multicollinearity.

We are going to use the code generated by:
https://seaborn.pydata.org/index.html

You don't need to understand each line. Only copy the code and change the name of the dataset for the one you are using.

```python
# Numpy: used for large, multi-dimensional arrays and matrices, and for high-level mathematical functions
# Pandas: used for data manipulation and analysis
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Dependent variable (bone mineral density of the femoral neck). In first column
# Independent variables (physical activity, body fat percentage, weight, and the interaction between weight and body fat) 
dataset = sns.load_dataset('titanic')
dataset.head()
```

```python
corr = dataset.corr()
plt.figure(figsize = (16,10))
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True,
    annot=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
```

As we can see, we don't have to worry about multi-collinearity in this dataset

```python
# However, check the next dataset
dataset2 = sns.load_dataset('car_crashes')
dataset2.head()
```

```python
corr = dataset2.corr()
plt.figure(figsize = (16,10))
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True,
    annot=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
```

We can see that total and no_previous features have a positive correlation over 0.96. Then, we could remove one of these features.
